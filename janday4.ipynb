## ðŸ”¹ What is a Machine Learning Model?

A Machine Learning model is a **learned rule**.

- It learns from past data
- It predicts future outcomes

**Model = relationship between input and output**


## ðŸ”¹ Supervised Learning

- Data is provided with correct answers
- Model learns from examples

**Our project:** Student Placement Prediction
- Input (Feature): CGPA
- Output (Target): Placement Status


```python
# Import required libraries
import numpy as np
```

## ðŸ”¹ Dataset Preparation

We define input feature and output target.

- 0 = Not Placed
- 1 = Placed


```python
cgpa = np.array([6.5, 7.0, 7.8, 8.2])
placed = np.array([0, 1, 1, 1])

print("CGPA:", cgpa)
print("Placed:", placed)
```

## ðŸ”¹ Reshaping Data for ML Model

ML models expect input features in **2D format**.

Shape before reshape: (4,)
Shape after reshape: (4, 1)


```python
X = cgpa.reshape(-1, 1)  # Feature
y = placed               # Target

print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

## ðŸ”¹ Import Machine Learning Model

We use **Logistic Regression**, a classification algorithm.

- Used for Yes / No prediction
- Commonly used in industry


```python
from sklearn.linear_model import LogisticRegression
```

## ðŸ”¹ Training the ML Model

- `.fit()` is where learning happens
- Model finds pattern between CGPA and Placement


```python
model = LogisticRegression()
model.fit(X, y)

print("Model training completed")
```

## ðŸ”¹ Making Predictions

Now we test the model with a new student.


```python
new_cgpa = np.array([[7.2]])
prediction = model.predict(new_cgpa)

print("Prediction:", prediction)
```

## ðŸ”¹ Human-Readable Output


```python
if prediction[0] == 1:
    print("Student is likely to get placed")
else:
    print("Student may not get placed")
```

## ðŸ”¹ Model Accuracy

Accuracy tells us how well the model learned from data.


```python
accuracy = model.score(X, y)
print("Model Accuracy:", accuracy)
```
